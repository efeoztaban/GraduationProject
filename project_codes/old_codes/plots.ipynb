# Data Prep

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

### Albedo

```python
inFile = open("albedo_data.txt")

albedo_event_nums_list=[]
albedo_cloud_num_list=[]
albedo_x_pos_list=[]
albedo_y_pos_list=[]
albedo_z_pos_list=[]
albedo_energy_list=[]
albedo_vertex_list=[]

a = inFile.readline()

for line in inFile:
  elements = line.split(";")
  albedo_event_nums_list.append(float(elements[0]))
  albedo_cloud_num_list.append(float(elements[1]))
  albedo_x_pos_list.append(float(elements[2]))
  albedo_y_pos_list.append(float(elements[3]))
  albedo_z_pos_list.append(float(elements[4]))
  albedo_energy_list.append(float(elements[5]))
  albedo_vertex_list.append(float(elements[6]))
```

```python
len(albedo_event_nums_list)
```

```python
albedo_event_nums_list[0]
```

```python
prev_event_num = 1

albedo = []

x_sum = 0
y_sum = 0
z_sum = 0
eng_sum = 0

cloud_xpos = []
cloud_ypos = []
cloud_zpos = []
cloud_energy = []

for i in range(len(albedo_event_nums_list)):

  if albedo_event_nums_list[i] != prev_event_num:

    my_dict = {}
    my_dict["x_pos"] = x_sum/eng_sum
    my_dict["y_pos"] = y_sum/eng_sum
    my_dict["z_pos"] = z_sum/eng_sum
    my_dict["energy"] = eng_sum
    my_dict["vertex"] = albedo_vertex_list[i-1]
    my_dict["x_pos_clouds"] = cloud_xpos
    my_dict["y_pos_clouds"] = cloud_ypos
    my_dict["z_pos_clouds"] = cloud_zpos

    albedo.append(my_dict)

    x_sum = 0
    y_sum = 0
    z_sum = 0
    eng_sum = 0

    cloud_xpos = []
    cloud_ypos = []
    cloud_zpos = []
    cloud_energy = []


    x_sum += albedo_x_pos_list[i]*albedo_energy_list[i]
    y_sum += albedo_y_pos_list[i]*albedo_energy_list[i]
    z_sum += albedo_z_pos_list[i]*albedo_energy_list[i]
    eng_sum += albedo_energy_list[i]

    cloud_xpos.append(albedo_x_pos_list[i])
    cloud_ypos.append(albedo_y_pos_list[i])
    cloud_zpos.append(albedo_z_pos_list[i])
    cloud_energy.append(albedo_energy_list[i])



  else:

    x_sum += albedo_x_pos_list[i]*albedo_energy_list[i]
    y_sum += albedo_y_pos_list[i]*albedo_energy_list[i]
    z_sum += albedo_z_pos_list[i]*albedo_energy_list[i]
    eng_sum += albedo_energy_list[i]

    cloud_xpos.append(albedo_x_pos_list[i])
    cloud_ypos.append(albedo_y_pos_list[i])
    cloud_zpos.append(albedo_z_pos_list[i])
    cloud_energy.append(albedo_energy_list[i])


  prev_event_num = albedo_event_nums_list[i]

my_dict = {}
my_dict["x_pos"] = x_sum/eng_sum
my_dict["y_pos"] = y_sum/eng_sum
my_dict["z_pos"] = z_sum/eng_sum
my_dict["energy"] = eng_sum
my_dict["vertex"] = albedo_vertex_list[i-1]
my_dict["x_pos_clouds"] = cloud_xpos
my_dict["y_pos_clouds"] = cloud_ypos
my_dict["z_pos_clouds"] = cloud_zpos

albedo.append(my_dict)
```

```python
lower_x = []
lower_y = []

limit = 5.0
limit2 = 4.5
```

### Albedo

```python
lower_x_pos_al = []
lower_y_pos_al = []

for i in albedo:
  if limit2 <= i["z_pos"] <= limit:
    lower_x_pos_al.append(i["x_pos"])
    lower_y_pos_al.append(i["y_pos"])

lower_x += lower_x_pos_al
lower_y += lower_y_pos_al
```

```python
font_size = 24

plt.figure(num=None, figsize=(15, 10), dpi=80)

plt.scatter(lower_x_pos_al, lower_y_pos_al,marker=".")

plt.ylabel("Y position (mm)")
plt.xlabel("X position (mm)")

plt.legend(prop={"size":font_size})
plt.show()
```

### CXB

```python
font_size = 24

plt.figure(num=None, figsize=(15, 10), dpi=80)

plt.scatter(lower_x_pos_cxb, lower_y_pos_cxb,marker=".")

plt.ylabel("Y position (mm)")
plt.xlabel("X position (mm)")

plt.legend(prop={"size":font_size})
plt.show()
```
